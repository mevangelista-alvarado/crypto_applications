## Crifrado César

Es un tipo de cifrado por sustitución en el que una letra en el texto original es reemplazada por otra letra que se encuentra un número fijo de posiciones más adelante en el alfabeto

image.png

Alfabeto base

```python
abc = 'abcdefghijklmnñopqrstuvwxyz'
```

```python
len(abc)
```

#### Alfabeto codificado

Hacemos la asignación $A = 0, B = 1, ..., Z = 27$

```python
key = 3
```

```python
new_abc = ""
for letra in abc:
  suma = abc.find(letra) + key
  modulo = int(suma) % len(abc)
  new_abc = new_abc + str(abc[modulo])

new_abc
```

La codificación anterior se puede describir matemáticamente como
$${\displaystyle x_{encriptada} = (x + desplazamiento)\mod {len(alfabelto)}}$$


#### Alfabeto Decodificado

```python
_abc = ''
for letra in new_abc:
    suma = new_abc.find(letra) - key
    modulo = int(suma) % len(new_abc)
    _abc = _abc + str(new_abc[modulo])
_abc
```

La decodificación anterior se puede describir matemáticamente como
$${\displaystyle x = (x_{encriptada} - desplazamiento)\mod {len(alfabelto)}}$$

##### Ahora en una función de Python.

```python
mensaje = "Hola Mundo"
key = 3
```

```python
def cifrar_palabra(cadena, clave):
    cadena = cadena.lower()
    text_cifrado = ''
    for letra in cadena:
        suma = abc.find(letra) + clave
        modulo = int(suma) % len(abc)
        text_cifrado = text_cifrado + str(abc[modulo])
    return text_cifrado

def decifrar_palabra(cadena, clave):
    text_cifrado = ''
    for letra in cadena:
        suma = abc.find(letra) - clave
        modulo = int(suma) % len(abc)
        text_cifrado = text_cifrado + str(abc[modulo])
    return text_cifrado
```

Lo anterior solo cifra palabras, ya que en mi alfabeto no cuento con otros símbolos

```python
def cifrar_mensaje(mensaje , clave):
  palabras = mensaje.split(" ")
  mensaje_cifrado = ""
  for palabra in palabras:
    _cifrar_palabra = cifrar_palabra(palabra, clave)
    mensaje_cifrado = f"{mensaje_cifrado} {_cifrar_palabra}"
  return mensaje_cifrado.strip()
```

```python
mensaje_cifrado = cifrar_mensaje(mensaje, key)
mensaje_cifrado
```

```python
def decifrar_mensaje(mensaje, clave):
  palabras = mensaje.split(" ")
  mensaje_decifrado = ""
  for palabra in palabras:
    _decifrar_palabra = decifrar_palabra(palabra, clave)
    mensaje_decifrado = f"{mensaje_decifrado} {_decifrar_palabra}"
  return mensaje_decifrado.strip()
```

```python
mensaje_decifrado = decifrar_mensaje(mensaje_cifrado, key)
mensaje_decifrado
```
